```
import sqlite3
conn = sqlite3.connect('/content/chinook.db') 
```

```
import pandas as pd
```

```
# Which Country preferes which genre?
cur = conn.cursor()
cur.execute(""" 
            SELECT invs.BillingCountry AS Countries,
            Round(SUm(invs.Total)) AS Revenue ,
	        	G.Name As Genre
            FROM invoices invs
            INNER JOIN invoice_items inv_it
            ON inv_it.InvoiceId = invs.InvoiceId
            INNER JOIN tracks T
            ON inv_it.TrackId = T.TrackId
            INNER JOIN genres G 
	          ON T.GenreId = G.GenreId
            GROUP BY  BillingCountry
            ORDER BY  Revenue Desc;
            """)
Genre_Country = cur.fetchall()
colnames = [desc[0] for desc in cur.description] 
conn.commit()


#create the pandas dataframe
Genre_Country = pd.DataFrame(Genre_Country)
Genre_Country.columns = colnames

#close the connection
cur.close()
```
